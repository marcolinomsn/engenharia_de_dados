#### Dados de estações representativas
O dataset conta com 26 anos de dados históricos para todas as estações do Brasil. O tamanho total do dataset é de 2 GB. Para facilitar o projeto do MVP e garantir uma representatividade em todo o territórios Brasileiro, foi feito um ranking das estações com maior período operacional e o histórico utilizado no projeto conta apenas com os dados observados dessas estações.



```python
# Defina o caminho do volume
volume_path = '/Volumes/mvp/staging/dataset'

# Crie o volume se necessário
dbutils.fs.mkdirs(volume_path)

# URL do CSV no GitHub
csv_url = 'https://raw.githubusercontent.com/marcolinomsn/engenharia_de_dados/refs/heads/main/dataset/weather_{}_filtered.csv'

# Caminho de destino no volume
csv_dest_path = volume_path + '/weather_{}_filtered.csv'

# Baixe o arquivo CSV para o volume
for year in range(2000, 2026):
    print(f"Baixando o arquivo CSV do ano de {year} para o volume")
    dbutils.fs.cp(csv_url.format(year), csv_dest_path.format(year))

# Leia o ultimo CSV usando Spark
df = spark.read.csv(csv_dest_path.format(year), header=True, inferSchema=True)

display(df.limit(10))
```
